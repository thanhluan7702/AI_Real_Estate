```python
df['Giá']=df['Giá'].astype(str)
```

```python
df['Giá'] = df['Giá'].apply(lambda x: 
    re.search(r"\d+\s*[.,]?\s*(\d*)?\s*(t[yỷiỉ]|tri[eêệẹ]u|tr|t)\s*(\d+(?:[.,]?(\d+)?)?)?\s*(t[yỷiỉ]|tri[eêệẹ]u|tr|t)?", x).group().replace(',', '.') 
    if re.search(r"\d+\s*(t[yỷiỉ]|tri[eêệẹ]u|tr|t)", x) else None)
```

```python
def trans_price(x):  
    if isinstance(x, str):  
        if (re.search('t[yỷiỉ]', x)) and (re.search('(tri[eêệẹ]u|tr)', x)):
            x = x.replace('ty', '.').replace('tỷ', '.').replace('ti', '.').replace('tỉ', '.').replace(',', '.').replace(' ','')
            match = re.search(r'\d+\s*\.\s*\d+', x)
            return float(match.group().strip()) * 1000000000 if match else None
        elif re.search('t[yỷiỉ]', x):
            if re.search('\d+\s*t[yỷiỉ]\s*\d+', x):
                x = x.replace('ty', '.').replace('tỷ', '.').replace('ti', '.').replace('tỉ', '.').replace(' ','')
                match = re.search(r'\d+\s*\.?\s*\d+', x)
                return float(match.group().strip()) * 1000000000 if match else None
            else:
                x = x.replace(',', '.').replace(' ','')
                match = re.search(r'\d*\s*\.?\s*\d+?', x)
                return float(match.group().strip()) * 1000000000 if match else None
        elif re.search('(tri[eêệẹ]u|tr)', x):
            if re.search('\d+\s*(tri[eêệẹ]u|tr)\s*\d+', x):
                x = x.replace('trieu', '.').replace('triệu', '.').replace('triêu', '.').replace('triẹu', '.').replace('tr', '.').replace(' ','')
                match = re.search(r'\d+\s*\.?\s*\d+', x)
                return float(match.group().strip()) * 1000000 if match else None
            else:
                x = x.replace(',', '.').replace(' ','')
                match = re.search(r'\d*\s*\.?\s*\d+?', x)
                return float(match.group().strip()) * 1000000 if match else None
    return None
```

```python
df['Giáa'] = df['Giá'].apply(lambda x: trans_price(x) if isinstance(x, str) and re.search(r"(t[yỷiỉ]|tri[eêệẹ]u|tr|t)", x) else None)
```
