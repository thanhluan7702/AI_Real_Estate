```python
import psycopg2
import os
# from datetime import date
import json
```

```python
def connect_to_database(database, user, password, host, port):
    try:
        connection = psycopg2.connect(
            database=database,
            user=user,
            password=password,
            host=host,
            port=port
        )
        print("Connected to the database")
        return connection
    except psycopg2.Error as e:
        print(f"Error: Unable to connect to the database\n{e}")
        return None

def close_connection(connection, cursor):
    if connection:
        connection.close()
        print("Connection closed")
    if cursor:
        cursor.close()
```

```python
def insert_nhatot_data(data, conn, cursor): 
    table_name = 'NHATOT'
    try:
        for row in data:
            columns = ', '.join(row.keys())
            values = ', '.join(['%s'] * len(row))
            insert_query = f"INSERT INTO {table_name} ({columns}) VALUES ({values})"
            cursor.execute(insert_query, tuple(row.values()))

        conn.commit()
        print("Data inserted successfully")
    except Exception as e:
        print(f"Error: Unable to insert data\n{e}")

def insert_batdongsan_data(data, conn, cursor):
    table_name = 'BATDONGSAN'
    try:
        for row in data:
            columns = ', '.join(['"' + key + '"' for key in row.keys()])
            values = ', '.join(['%s'] * len(row))
            insert_query = f'INSERT INTO {table_name} ({columns}) VALUES ({values}) ON CONFLICT DO NOTHING'
            cursor.execute(insert_query, tuple(row.values()))

        conn.commit()
        print("Data inserted successfully")
    except Exception as e:
        print(f"Error: Unable to insert data\n{e}")
        
def insert_alonhadat_data(data, conn, cursor): 
    table_name = 'ALONHADAT'
    try:
        for row in data:
            columns = ', '.join(row.keys())
            values = ', '.join(['%s'] * len(row))
            insert_query = f"INSERT INTO {table_name} ({columns}) VALUES ({values})"
            cursor.execute(insert_query, tuple(row.values()))

        conn.commit()
        print("Data inserted successfully")
    except Exception as e:
        print(f"Error: Unable to insert data\n{e}")
```
