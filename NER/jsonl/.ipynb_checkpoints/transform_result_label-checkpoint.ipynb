### 1. Transform

```python
import glob 
import jsonlines 
import pandas as pd 
```

```python
data = [] 

for file in glob.glob('data_0.jsonl'): 
    with jsonlines.open(file, 'r') as reader: 
        data += list(reader)   
```

```python
lst_text = []
lst_tags = [] 

for record in data: 
    text = record['document_text']
    lst_text.append(text)
    
    labels = [] 
    values = [] 

    for tag in record['labels']: 
        entity = tag['label_name']
        value = text[tag['span_start'] : tag['span_end']]
        
        for i, v in enumerate(value.split()):
            if i == 0: 
                labels.append(f'B-{entity}')
            else: 
                labels.append(f'I-{entity}') 
            values.append(v)
            
    for i, value in enumerate(values): 
        text = text.replace(value, labels[i])

    lst = text.split()

    for idx, ent in enumerate(lst): 
        if ent[:2] not in ('B-', 'I-'):  
            lst[idx] = 'O'
    lst_tags.append(' '.join(lst))
```

```python
df = pd.DataFrame({'text' : lst_text, 
              'tags' : lst_tags})

df.tags = df.tags.replace(',', '')
```

### 2. Check entity error

```python
check = {}

for tag in df.tags: 
    for t in tag.split(): 
        if t not in check: 
            check[t] = len(check)
```
