```python
df=pd.read_csv('data_1.csv')
```

```python
def extract_information(description):
    result = {
        'text_cleaned': description,
        'LOAI_BDS': None,
        'GIA': None,
        'DIA_CHI': None,
        'DIEN_TICH': None,
        'DAI_RONG': None,
        'HUONG_NHA': None,
        'VI_TRI': None,
        'DAN_CU': None,
        'VIA_HE': None,
        'DUONG_KIET': None,
        'VIEW': None,
        'HINH_DANG_NHA': None,
        'TIEN_ICH_KHAC': None,
        'SO_PHONG_NGU': None,
        'SO_TOILET': None,
        'SO_TANG': None,
        'NOI_THAT': None,
        'KV_DE_XE': None,
        'PHAP_LY': None,
        'DICH_VU': None,
        'AN_NINH': None
        # Th√™m c√°c tr∆∞·ªùng th√¥ng tin kh√°c n·∫øu c·∫ßn
    }

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t gi√°
    gia_pattern = re.compile(r'gi√°[^0-9]*(\d+(?:[.,]\d+)?)\s*(?:ty|ti|trieu|t·ª∑|t·ªâ|tri·ªáu|tr)?\s*(\d+(?:[.,]\d+)?)?\s*(?:ty|ti|trieu|t·ª∑|t·ªâ|tri·ªáu|tr)?\s*(\d+(?:[.,]\d+)?)?\s*(\/[m2|m¬≤])?', re.IGNORECASE)
    match_gia = gia_pattern.search(description)
    if match_gia:
        result['GIA'] = match_gia.group().strip() 

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t di·ªán t√≠ch
    dien_tich_pattern = re.compile(r'(di·ªán t√≠ch|dt|s|S|dtich|dt)[^0-9]*(\d+(?:[.,]\d+)?)\s*m2', re.IGNORECASE)
    match_dien_tich = dien_tich_pattern.search(description)
    if match_dien_tich:
        result['DIEN_TICH'] = match_dien_tich.group()
        
    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t k√≠ch th∆∞·ªõc
    dai_rong_pattern = re.compile(r'\(?\s*(\d+(?:[.,]\d+)?)\s*[x*-]\s*(\d+(?:[.,]\d+)?)\)?', re.IGNORECASE)
    match_dai_rong = dai_rong_pattern.search(description)
    if match_dai_rong:
        result['DAI_RONG'] = match_dai_rong.group()   

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t ƒë·ªãa ch·ªâ
    dia_chi_pattern = re.compile(r'ƒë·ªãa ch·ªâ[^.]*: (.+?)\.', re.IGNORECASE)
    match_dia_chi = dia_chi_pattern.search(description)
    if match_dia_chi:
        result['DIA_CHI'] = match_dia_chi.group()

        
    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t h∆∞·ªõng nh√†
    huong_nha_pattern = re.compile(r'(h∆∞·ªõng|h)\s*(ƒë√¥ng|nam|t√¢y|b·∫Øc)\s*(ƒë√¥ng|nam|t√¢y|b·∫Øc)?', re.IGNORECASE)
    match_huong_nha = huong_nha_pattern.search(description)
    if match_huong_nha:
        result['HUONG_NHA'] = match_huong_nha.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t v·ªã tr√≠
    vi_tri_pattern = re.compile(r'v·ªã tr√≠ (.+?)\.', re.IGNORECASE)
    match_vi_tri = vi_tri_pattern.search(description)
    if match_vi_tri:
        result['VI_TRI'] = match_vi_tri.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t d√¢n c∆∞
    dan_cu_pattern = re.compile(r'd√¢n c∆∞ (.+?)\s*[.,]', re.IGNORECASE)
    match_dan_cu = dan_cu_pattern.search(description)
    if match_dan_cu:
        result['DAN_CU'] = match_dan_cu.group()
        
    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t via he
    via_he_pattern = re.compile(r'(v·ªâa h√®|l·ªÅ)\s*(\d+(?:[.,]\d+)?)\s*m(\d+(?:[.,]\d+)?)?', re.IGNORECASE)
    match_via_he = via_he_pattern.search(description)
    if match_via_he:
        result['VIA_HE'] = match_via_he.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t ƒë∆∞·ªùng ki·ªát
    duong_kiet_pattern = re.compile(r'(ƒë∆∞·ªùng r·ªông|ki·ªát|ƒë∆∞·ªùng)\s*(\d+(?:[.,]\d+)?)\s*m(\d+(?:[.,]\d+)?)?', re.IGNORECASE)
    match_duong_kiet = duong_kiet_pattern.search(description)
    if match_duong_kiet:
        result['DUONG_KIET'] = match_duong_kiet.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t view
    view_pattern = re.compile(r'view\s*(.+?)[.,]', re.IGNORECASE)
    match_view = view_pattern.search(description)
    if match_view:
        result['VIEW'] = match_view.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t h√¨nh d·∫°ng nh√†
    hinh_dang_nha_pattern = re.compile(r'h√¨nh d√°ng nh√† (.+?)\.', re.IGNORECASE)
    match_hinh_dang_nha = hinh_dang_nha_pattern.search(description)
    if match_hinh_dang_nha:
        result['HINH_DANG_NHA'] = match_hinh_dang_nha.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t ti·ªán √≠ch kh√°c
    tien_ich_khac_pattern = re.compile(r'ti·ªán √≠ch kh√°c: (.+?)\.', re.IGNORECASE)
    match_tien_ich_khac = tien_ich_khac_pattern.search(description)
    if match_tien_ich_khac:
        result['TIEN_ICH_KHAC'] = match_tien_ich_khac.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t s·ªë ph√≤ng ng·ªß
    so_phong_ngu_pattern = re.compile(r'(\d+)\s*(ph√≤ng ng·ªß|pn)', re.IGNORECASE)
    match_so_phong_ngu = so_phong_ngu_pattern.search(description)
    if match_so_phong_ngu:
        result['SO_PHONG_NGU'] = match_so_phong_ngu.group()
        
    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t s·ªë toilet
    so_toilet_pattern = re.compile(r'(\d+)\s*(toilet|ph√≤ng v·ªá sinh|ph√≤ng vs|vs|pvs|wc)', re.IGNORECASE)
    match_so_toilet = so_toilet_pattern.search(description)
    if match_so_toilet:
        result['SO_TOILET'] = match_so_toilet.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t s·ªë t·∫ßng
    so_tang_pattern = re.compile(r'(\d+)\s*t·∫ßng', re.IGNORECASE)
    match_so_tang = so_tang_pattern.search(description)
    if match_so_tang:
        result['SO_TANG'] = match_so_tang.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t n·ªôi th·∫•t
    noi_that_pattern = re.compile(r'n·ªôi th·∫•t (.+?)[.,]', re.IGNORECASE)
    match_noi_that = noi_that_pattern.search(description)
    if match_noi_that:
        result['NOI_THAT'] = match_noi_that.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t khu v·ª±c ƒë·ªÉ xe
    kv_de_xe_pattern = re.compile(r'khu v·ª±c ƒë·ªÉ xe (.+?)[.,]', re.IGNORECASE)
    match_kv_de_xe = kv_de_xe_pattern.search(description)
    if match_kv_de_xe:
        result['KV_DE_XE'] = match_kv_de_xe.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t th√¥ng tin ph√°p l√Ω
    phap_ly_pattern = re.compile(r'ph√°p l√Ω\s*(.+?)[.,]', re.IGNORECASE)
    match_phap_ly = phap_ly_pattern.search(description)
    if match_phap_ly:
        result['PHAP_LY'] = match_phap_ly.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t th√¥ng tin d·ªãch v·ª•
    dich_vu_pattern = re.compile(r'd·ªãch v·ª• (.+?)[.,]', re.IGNORECASE)
    match_dich_vu = dich_vu_pattern.search(description)
    if match_dich_vu:
        result['DICH_VU'] = match_dich_vu.group()

    # Bi·ªÉu th·ª©c ch√≠nh quy ƒë·ªÉ tr√≠ch xu·∫•t th√¥ng tin an ninh
    an_ninh_pattern = re.compile(r'an ninh\s*(.+?)[.,]', re.IGNORECASE)
    match_an_ninh = an_ninh_pattern.search(description)
    if match_an_ninh:
        result['AN_NINH'] = match_an_ninh.group()


    return result
```

```python
df_results = pd.DataFrame([extract_information(row['text_cleaned']) for _, row in df.iterrows()])
```
